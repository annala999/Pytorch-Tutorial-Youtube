## Random tensors
### Why random tensors?
#### Random tensor are important because the way many networks learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data.
##### Start with random numbers -> look at data -> updata random numbers -> look at data -> updata random *numbers*

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(3,4)
random_tensor
```

```python
# Create a random tensor with similar shape to an image tensor
random_image_size_tensor = torch.rand(size=(3,244,244)) #height,width,color channels (R,G,B)
random_image_size_tensor.shape,random_image_size_tensor.ndim
```

```python
random_image_size_tensor = torch.rand(size=(244,244,3)) #height,width,color channels (R,G,B)
random_image_size_tensor.shape,random_image_size_tensor.ndim
```

## Tensor datatypes

**Note:** Tensor datatype is one of the 3 big errors you'll run into with PyTorch & deep learning:

1.Tensors not right datatype

2.Tensors not right shape

3.Tensors not on the right device


```python
# Float 32 tensor
float_32_tensor = torch.tensor([3.0,6.0,9.0],
                               dtype=None,   #What datatype is the tensor (e.g. float32 or float16)
                               device=None,  # What device is your tensor on
                               requires_grad=False) # Whether or not to track gradients with this tensors operations
float_32_tensor
```

```python
float_32_tensor.dtype
```

```python
non_default_tensor = torch.tensor([3.0,6.0,9.0],
                                  dtype=torch.float16)
non_default_tensor,non_default_tensor.dtype
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_16_tensor*float_32_tensor
```

```python
int_32_tensor = torch.tensor([3,6,9],dtype=torch.int64)
int_32_tensor
```

```python
int_32_tensor = torch.tensor([3,6,9],dtype=torch.long)
int_32_tensor
```

### Getting information from tensors ( tensor attributes)
1.Tensors not right datatype - to do get datatype from a tensor,can use tensor.dtype

2.Tensors not right shape - to get shape from a tensor,can use tensor.shape

3.Tensors not on the right device - to get device from a tensor,can use tensor.device

```python
# Create a tensor
some_tensor = torch.rand(3,4)
some_tensor
```

```python
some_tensor.size,some_tensor.shape
```

```python
# Find out details about some tensor
print(some_tensor)
print(f"Datatype of tensor:{some_tensor.dtype}")
print(f"Shape of tensor:{some_tensor.shape}")
print(f"Device tensor is on:{some_tensor.device}")
```
